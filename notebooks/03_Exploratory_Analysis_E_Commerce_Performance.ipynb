```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ecommerce_data = pd.read_csv("/Users/yssf/Desktop/E-Commerce-Performance-Analysis-main/data/e-commerce-dataset.csv")
clickstream_data = pd.read_csv("/Users/yssf/Desktop/E-Commerce-Performance-Analysis-main/data/e-commerce_clickstream_transactions.csv")
```

```python
print(ecommerce_data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=clickstream_data, x='EventType', order=clickstream_data['EventType'].value_counts().index)
plt.title('Type d\'événements par type')
plt.xlabel('Type d\'événement')
plt.ylabel('Nombre d\'événements')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(clickstream_data['Amount'].dropna(), bins=30, kde=True)
plt.title('Répartition des montants des achats')
plt.xlabel('Montant')
plt.ylabel('Fréquence')
plt.show()
```

```python
clickstream_data['Timestamp'] = pd.to_datetime(clickstream_data['Timestamp'])
clickstream_data['Date'] = clickstream_data['Timestamp'].dt.date
```

```python
daily_events = clickstream_data['Date'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
daily_events.plot(kind='line')
plt.title('Événements quotidiens')
plt.xlabel('Date')
plt.ylabel('Nombre d\'événements')
plt.xticks(rotation=45)
plt.show()
```

```python
clickstream_data['Amount'].dropna().astype(float).hist(bins=30, edgecolor='black')
plt.title('Répartition des montants associés aux événements')
plt.xlabel('Montant')
plt.ylabel('Fréquence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ecommerce_data['Time on Site'], bins=30, kde=True)
plt.title('Répartition du temps passé sur le site')
plt.xlabel('Temps sur le site (en secondes)')
plt.ylabel('Fréquence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ecommerce_data['Age'], bins=10, kde=True)  # Utiliser l'âge
plt.title('Réparation de l\'âge des clients')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ecommerce_data['Annual Income'], bins=30, kde=True)
plt.title('Analyse du revenu annuel')
plt.xlabel('Revenu annuel')
plt.ylabel('Fréquence')
plt.show()
```

** Explanations of the stages: **

- types of events by site: how users interact with the site,
- Distribution of purchase amounts
- Analysis of daily events over time that can reveal trends in user behavior
- distribution of amounts associated with events
- Time distribution spent on the site (seconds)
- Distribution of customer age
- Analysis of customer annual income
