```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ecommerce_data = pd.read_csv("/Users/yssf/Desktop/E-Commerce-Performance-Analysis-main/data/e-commerce-dataset.csv")
clickstream_data = pd.read_csv("/Users/yssf/Desktop/E-Commerce-Performance-Analysis-main/data/e-commerce_clickstream_transactions.csv")
```

```python

plt.figure(figsize=(15, 8))
sns.histplot(data=ecommerce_data, x='Annual Income', bins=30, kde=True)
plt.title('Distribution du revenu annuel')
plt.xlabel('Revenu annuel')
plt.ylabel('Fr√©quence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=ecommerce_data, x='Gender')
plt.title('R√©partition des clients par genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de clients')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=ecommerce_data, x='Age', y='Annual Income', hue='Gender', alpha=0.6)
plt.title('Relation entre l\'√¢ge et le revenu annuel')
plt.xlabel('√Çge')
plt.ylabel('Revenu annuel')
plt.legend(title='Genre')
plt.show()
```

The analysis of e-commerce data allowed me to explore key information on customer behavior and their relationship with income.The visualizations made revealed interesting trends:

** Distribution of annual income: **
- The distribution of annual income shows peaks at certain income levels, which can indicate segments of specific customers that could be targeted by marketing campaigns.

** Distribution by genre: **
- The distribution of customers by genre highlights the differences in the customer base, which can help to orient marketing and product strategies to better meet the expectations of each segment.

** The relationship between age and annual income: **
- The visualization of the relationship between age and annual income suggests that there are specific age segments with potentially higher income that can influence product development and advertising targeting decisions.

In conclusion, these insights can help the company ** to refine its marketing strategies, to personalize offers for different customer segments and to improve the user experience. **
